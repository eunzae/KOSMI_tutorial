```python
%%R
dat$aline_flg<-as.factor(dat$aline_flg)
dat$gender_num<-as.factor(dat$gender_num)
dat$service_num<-as.factor(dat$service_num)
dat$hosp_exp_flg<-as.factor(dat$hosp_exp_flg)
dat$icu_exp_flg <-as.factor(dat$icu_exp_flg )
dat$day_28_flg<-as.factor(dat$day_28_flg)
dat$censor_flg<-as.factor(dat$censor_flg)
dat$sepsis_flg<-as.factor(dat$sepsis_flg)
dat$chf_flg<-as.factor(dat$chf_flg)
dat$afib_flg<-as.factor(dat$afib_flg)
dat$renal_flg<-as.factor(dat$renal_flg)
dat$liver_flg<-as.factor(dat$liver_flg)
dat$copd_flg<-as.factor(dat$copd_flg)
dat$cad_flg<-as.factor(dat$cad_flg)
dat$stroke_flg<-as.factor(dat$stroke_flg)
dat$mal_flg<-as.factor(dat$mal_flg)
dat$resp_flg<-as.factor(dat$resp_flg)
```

```python
%%R
datby.aline <- split(dat,dat$aline_flg)
tab2 <- cbind(produce.table1(datby.aline[[1]]),
              produce.table1(datby.aline[[2]]))
colnames(tab2) <- c("No-IAC", "IAC")
kable(tab2, caption="Patient characteristics stratified by IAC administration")
```

```python
%%R
co2_lm <- lm(tco2_first ~ pco2_first,data=dat)
plot(formula=tco2_first~pco2_first, data=dat, xlab="PCO2", ylab="TCO2", pch=19)
abline(co2_lm,col='red',lwd=2)
```

```python
%%R
summary(co2_lm)
```

```python
%%R
dat$gender_num <- as.factor(dat$gender_num)
co2_gender_lm <- lm(tco2_first ~ pco2_first + gender_num,data=dat)
summary(co2_gender_lm)
```

```python
%%R
confint(co2_lm)
```

```python
%%R
confint(co2_gender_lm)
```

```python
%%R
grid_pred <- data.frame(pco2_first=seq.int(from=min(dat$pco2_first,na.rm=T),
                                           to=max(dat$pco2_first,na.rm=T)));
head(grid_pred,5)
```

```python
%%R
preds <- predict(co2_lm,newdata=grid_pred,interval = "prediction")
preds[1:5,]
```

```python
%%R
plot(dat$pco2_first,dat$tco2_first,xlab="PCO2",ylab="TCO2",pch=19)
abline(co2_lm,col='red',lwd=2)
lines(grid_pred$pco2_first,preds[,2],lty=3)
lines(grid_pred$pco2_first,preds[,3],lty=3)
```

```python
%%R
dat$mort_day_censored[dat$censor_flg==1] <- 731
datSurv <- Surv(dat$mort_day_censored,dat$censor_flg==0)
datSurv[101:105]
```

```python
%%R
gender_surv <- survfit(datSurv~gender_num,data=dat)
plot(gender_surv,col=1:2,conf.int = TRUE,xlab="Days",ylab="Proportion Who Survived")
legend(500,0.2,col=1:2,lty=1,c("Women","Men"))
```

```python
%%R
unit_surv <- survfit(datSurv~service_unit,data=dat)
plot(unit_surv,col=1:3,conf.int = FALSE,xlab="Days",ylab="Proportion Who Survived")
legend(500,0.2,col=1:3,lty=1,c("FICU","MICU","SICU"))
```

```python
%%R
gender_coxph <- coxph(datSurv ~ gender_num,data=dat)
summary(gender_coxph)
```
